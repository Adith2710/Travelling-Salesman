```
import pandas as pd
import numpy as np

import math
```

```
#Importing all the data from CSV
#Copy the path of the CSV file and paste it here

raw_data=pd.read_csv("/Kiosk Coords (CSV).csv")
raw_data

```

```
#cols=raw_data.shape[1]
#takes in the lattitude as X

x_raw=np.matrix((raw_data.iloc[:,2]))

#takes longitude as as Y
y_raw=np.matrix(raw_data.iloc[:,3])

#an array for storing the locations 
location=np.array(raw_data.iloc[:,0])

x=x_raw
y=y_raw


```

```
# function to find distance between two points

def dist(x1, x2, y1, y2):
    distance = math.sqrt((x2-x1)**2 + (y2-y1)**2)

    return float(distance)
```

```
# initializing an adjacency matrix of size (50,50) and setting it the values with distances with coressponding points. diagonal points are infinity.
adj_m= np.zeros(((x.shape[1]),(x.shape[1])))


for i in range(x.shape[1]):
  for j in range(x.shape[1]):
    
    
    if i==j:
      
      adj_m[i,j]=math.inf
    else:
      adj_m[i,j]=dist(x[0,i],x[0,j],y[0,i],y[0,j])
      
#print(adj_m)
```

```
#from this adjacency matrix, we start from 0th row and find the index of the minimum value of that row, i.e the shortest path
# now that index is saved in "min_val_in". inital point is is "point". the columns of the intial point and min_val_in 
# are set to infinity so that no point can return back.
# the min_val_in is saved in a list ans
# now the intial point is set to min_val_in and the loop is run again. it is done till  all 50 points are covered

count=0
ans=[0]
point=0
while(count<x.shape[1]):
  min_val_in=np.argmin(adj_m[point])
  ans.append(min_val_in)
  adj_m[:,point]=adj_m[:,min_val_in]=math.inf
  point=min_val_in
  count+=1
  
#print(ans)
```

```
# the shortest path is in ans, and that is made to display with its corresponding location with the dictionary 'route'

route={i+1:location[ans[i]] for i in range(len(ans)) }
route
```
